```python
Retail = pd.read_csv("OnlineRetail.csv")
Retail
```

```python
fin_Retail = Retail.drop_duplicates(['CustomerID','Description'])
fin_Retail
```

```python
fin_Retail.shape
```

```python
pivot = fin_Retail.pivot_table(columns='CustomerID',index='Description',values='UnitPrice')
pivot
```

```python
pivot.shape
```

```python
pivot.fillna(0,inplace=True)
pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
retail_sparse = csr_matrix(pivot)
```

```python
retail_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(retail_sparse)
```

```python
distance, suggestion = model.kneighbors(pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('ORRS/model.pkl','wb'))
pickle.dump(name, open('ORRS/name.pkl','wb'))
pickle.dump(fin_Retail, open('ORRS/fin_Retail.pkl','wb'))
pickle.dump(pivot, open('ORRS/pivot.pkl','wb'))
```

```python
def recom_retail(retail_name):
    retail_id = np.where(pivot.index == retail_name)[0][0]
    distance, suggestion = model.kneighbors(pivot.iloc[retail_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
        retails = pivot.index[suggestion[i]]
        for j in retails:
            print(j)
```

```python
retail_name = "ASSTD COL BUTTERFLY/CRYSTAL W/CHIME"
recom_retail(retail_name)
```
